```python
import os
import cv2
```

```python
# img_path = '/local_hdd/works/ICRyu_workspace/for_yolo/dataset/data2/image_2'
# img_path = '/local_hdd/works/ICRyu_workspace/for_yolo/dataset/data1/target_label_for_NH/1~15000'
img_path = '/local_hdd/works/ICRyu_workspace/for_yolo/dataset/data1/target_label_for_NH/15001~30000'
save_path = '/local_hdd/works/ICRyu_workspace/for_yolo/images'
```

```python
matching_list = []
for root, subs, files in os.walk(img_path):
    for file in files:
        filename = os.path.splitext(file)[0]
        ext = os.path.splitext(file)[-1]
        if ext == '.png':
            origin_full_path = os.path.join(root, file)
            new_save_path = os.path.join(save_path, filename + '.jpg')
            matching_tmp = [origin_full_path, new_save_path]
            print(matching_tmp)
            matching_list.append(matching_tmp)
```

```python
for each_list in matching_list:
    origin_path = each_list[0]
    save_path = each_list[1]

    arr = cv2.imread(origin_path)
    cv2.imwrite(save_path, arr)
```

```python
img_list = os.listdir('/local_hdd/works/ICRyu_workspace/for_yolo/images')
tmp = []
for elem in img_list:
    ext = os.path.splitext(elem)[-1]
    if ext in ['.jpg', '.png']:
        file_path = os.path.join('data/obj', elem)
        tmp.append(file_path)

print(tmp)
```

```python
with open('/local_hdd/works/ICRyu_workspace/for_yolo/train.txt', 'w') as f:
    for i in tmp:
        f.write(f'{i}')
        f.write('\n')
```
